```python
plt.style.use("bmh")
sns.rugplot(df["total_bill"])
```

```python
# Comparing the different kinds of distribution plots  using subplot
# Histogram
plt.style.use('bmh')
plt.figure(figsize=(12, 6))
plt.subplot(2, 2, 2)
sns.histplot(df['total_bill'], kde=True)
plt.title('Histogram with KDE')

# KDE Plot
plt.subplot(2, 2, 1)
sns.kdeplot(df['total_bill'])
plt.title('KDE Plot')

# Rug Plot
plt.subplot(2, 2, 3)
sns.rugplot((df['total_bill']))
plt.title('Rug Plot')


# Rug Plot
plt.subplot(2, 2, 4)
sns.ecdfplot(df['total_bill'])
plt.title('ECDF Plot')
plt.tight_layout()
plt.show()
```

```python
plt.style.use('bmh')
sns.histplot(df['tip'], bins=5)
```

```python
plt.subplot(2, 1, 2)
sns.displot(df['tip'])
plt.title('dis Plot')
plt.tight_layout()
plt.show()

plt.subplot(1, 2, 1)
sns.kdeplot(df['tip'])
plt.title('kde Plot')
plt.tight_layout()
plt.show()
```
